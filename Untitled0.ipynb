```python
import matplotlib.pyplot as plt
import numpy as np

# 1. Menentukan rentang nilai n (dari 0 sampai 30)
n = np.arange(0, 31)  # np.arange(0, 31) akan menghasilkan angka 0, 1, 2, ..., 30

# 2. Membuat Sinyal Pertama: x1[n] = sin(0.5 * pi * n)
x1 = np.sin(0.5 * np.pi * n)

# 3. Membuat Sinyal Kedua: Unit Step (x2[n] = 0 untuk n < 5, dan 1 untuk n >= 5)
# np.where(kondisi, nilai_jika_benar, nilai_jika_salah)
x2 = np.where(n < 5, 0, 1)

# 4. Visualisasi Sinyal x1[n]
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
plt.stem(n, x1, basefmt="C2-")  # plt.stem digunakan khusus untuk sinyal diskrit (grafik jarum)
plt.title(r"Sinyal Pertama: $x_1[n] = \sin(0.5\pi n)$")
plt.xlabel("Sampel (n)")
plt.ylabel("Amplitudo")
plt.grid(True)

# 5. Visualisasi Sinyal x2[n]
plt.stem(n, x2, basefmt="C2-")
plt.title(r"Sinyal Kedua: Unit Step $x_2[n]$ ($n \geq 5$)")
plt.xlabel("Sampel (n)")
plt.ylabel("Amplitudo")
plt.grid(True)

# Menampilkan semua plot
plt.tight_layout()
plt.show()
```

```python
# Lanjutan dari cell pertama, variabel n, x1, dan x2 masih tersimpan di memori

# 1. Melakukan operasi penjumlahan sinyal
y = x1 + x2

# 2. Membuat visualisasi (3 plot bertingkat agar mudah dibandingkan)
plt.figure(figsize=(10, 12))

# Plot x1[n]
plt.subplot(3, 1, 1)
plt.stem(n, x1, linefmt="b-", markerfmt="bo", basefmt="C2-")
plt.title(r"Sinyal Asal 1: $x_1[n] = \sin(0.5\pi n)$")
plt.ylabel("Amplitudo")
plt.grid(True)

# Plot x2[n]
plt.subplot(3, 1, 2)
plt.stem(n, x2, linefmt="r-", markerfmt="ro", basefmt="C2-")
plt.title(r"Sinyal Asal 2: Unit Step $x_2[n]$")
plt.ylabel("Amplitudo")
plt.grid(True)

# Plot Hasil Penjumlahan y[n]
plt.subplot(3, 1, 3)
plt.stem(n, y, linefmt="g-", markerfmt="go", basefmt="C2-")
plt.title(r"Sinyal Hasil Penjumlahan: $y[n] = x_1[n] + x_2[n]$")
plt.xlabel("Sampel (n)")
plt.ylabel("Amplitudo")
plt.grid(True)

# Menampilkan semua plot
plt.tight_layout()
plt.show()
```

```python
# Lanjutan dari cell sebelumnya, variabel n dan x1 masih tersimpan di memori
x_asli = x1  # Kita gunakan sinyal x1[n] = sin(0.5 * pi * n) sebagai objek eksperimen

# 1. Melakukan operasi penggeseran sinyal dengan memanipulasi indeks n
# y[n] = x[n - k]
# Jika kita menggeser sinyal, kita bisa membuat fungsi bantu untuk menangani indeksnya


def geser_sinyal(n_array, x_array, k):
    # Membuat array kosong berukuran sama dengan n penuh dengan angka 0
    y_geser = np.zeros_like(x_array)

    for i, ni in enumerate(n_array):
        indeks_asal = ni - k
        # Memastikan indeks asal masih berada di dalam rentang 0 sampai 30
        if 0 <= indeks_asal <= 30:
            y_geser[i] = x_array[indeks_asal]
    return y_geser


# Menghitung hasil pergeseran untuk masing-masing k
y_k_neg3 = geser_sinyal(n, x_asli, -3)  # k = -3
y_k_0 = geser_sinyal(n, x_asli, 0)  # k = 0 (tetap)
y_k_pos4 = geser_sinyal(n, x_asli, 4)  # k = 4

# 2. Membuat visualisasi perbandingan dalam satu gambar (4 plot bertingkat)
plt.figure(figsize=(10, 14))

# Plot k = -3 (Maju / Advance)
plt.subplot(4, 1, 1)
plt.stem(n, y_k_neg3, linefmt="m-", markerfmt="mo", basefmt="C2-")
plt.title("Sinyal Hasil Penggeseran $k = -3$ ($y[n] = x[n + 3]$)")
plt.ylabel("Amplitudo")
plt.grid(True)

# Plot k = 0 (Sinyal Asli)
plt.subplot(4, 1, 2)
plt.stem(n, y_k_0, linefmt="b-", markerfmt="bo", basefmt="C2-")
plt.title("Sinyal Asli $k = 0$ ($y[n] = x[n]$)")
plt.ylabel("Amplitudo")
plt.grid(True)

# Plot k = 4 (Mundur / Delay)
plt.subplot(4, 1, 3)
plt.stem(n, y_k_pos4, linefmt="r-", markerfmt="ro", basefmt="C2-")
plt.title("Sinyal Hasil Penggeseran $k = 4$ ($y[n] = x[n - 4]$)")
plt.ylabel("Amplitudo")
plt.grid(True)

# Plot Perbandingan Overlay (Semua jadi satu untuk melihat pergeseran horizontal)
plt.subplot(4, 1, 4)
plt.plot(n, y_k_0, "b-o", label="Asli (k=0)", alpha=0.5)
plt.plot(n, y_k_neg3, "m--v", label="Maju (k=-3)", alpha=0.7)
plt.plot(n, y_k_pos4, "r--x", label="Tunda (k=4)", alpha=0.7)
plt.title("Perbandingan Seluruh Hasil Penggeseran")
plt.xlabel("Sampel (n)")
plt.ylabel("Amplitudo")
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()
```

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np

# 1. Membaca Citra 1 (Foto Lubang Kunci Kamu)
citra1_bgr = cv2.imread("/fotoku.jpg")

if citra1_bgr is None:
    print(
        "ERROR: File '/fotoku.jpg' tidak ditemukan. Pastikan path sudah benar!"
    )
else:
    # Ubah ke Grayscale
    citra1 = cv2.cvtColor(citra1_bgr, cv2.COLOR_BGR2GRAY)
    tinggi, lebar = citra1.shape

    # 2. Membuat Citra 2: Sinyal Pola Gradasi (Sengaja dibuat ukuran berbeda untuk simulasi resizing)
    # Kita buat ukuran aslinya sengaja berbeda, misal 300 x 300
    citra2_asal = np.zeros((300, 300), dtype=np.uint8)
    for i in range(300):
        citra2_asal[:, i] = int((i / 300) * 255)  # Pola gradasi dari gelap ke terang

    # 3. PROSES RESIZING (Wajib Tugas Poin 4)
    # Mengubah ukuran Citra 2 agar pas dengan ukuran Citra 1 (tinggi x lebar)
    citra2 = cv2.resize(citra2_asal, (lebar, tinggi), interpolation=cv2.INTER_LINEAR)

    # 4. OPERASI PENJUMLAHAN CITRA
    # Metode A: Penjumlahan Biasa dengan Saturation/Clipping (Bawaan OpenCV)
    # Jika hasil > 255, otomatis dikunci di angka 255
    citra_hasil_clip = cv2.add(citra1, citra2)

    # Metode B: Penjumlahan dengan Normalisasi (Menggunakan NumPy Float)
    # Hasil dijumlahkan dulu apa adanya, lalu diskalakan kembali ke rentang 0-255
    citra_jumlah_float = citra1.astype(float) + citra2.astype(float)
    citra_hasil_norm = cv2.normalize(
        citra_jumlah_float, None, 0, 255, cv2.NORM_MINMAX
    ).astype(np.uint8)

    # 5. MENAMPILKAN VISUALISASI OUTPUT
    plt.figure(figsize=(14, 10))

    # Plot Citra 1
    plt.subplot(2, 2, 1)
    plt.imshow(citra1, cmap="gray")
    plt.title(f"Citra 1: Foto Asli ({lebar}x{tinggi})")
    plt.axis("off")

    # Plot Citra 2
    plt.subplot(2, 2, 2)
    plt.imshow(citra2, cmap="gray")
    plt.title(f"Citra 2: Hasil Resize Gradasi ({lebar}x{tinggi})")
    plt.axis("off")

    # Plot Hasil Penjumlahan Metode Clipping
    plt.subplot(2, 2, 3)
    plt.imshow(citra_hasil_clip, cmap="gray")
    plt.title("Hasil Penjumlahan (Metode Clipping/Saturation)")
    plt.axis("off")

    # Plot Hasil Penjumlahan Metode Normalisasi
    plt.subplot(2, 2, 4)
    plt.imshow(citra_hasil_norm, cmap="gray")
    plt.title("Hasil Penjumlahan (Metode Normalisasi)")
    plt.axis("off")

    plt.tight_layout()
    plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Menggunakan kembali n dan x1 (sinus) dari Bagian A sebagai input x
n = np.arange(0, 31)
x = np.sin(0.5 * np.pi * n)


# 1. Mendefinisikan Fungsi Sistem T(x) = 2 * x
def sistem_T(sinyal_input):
    return 2 * sinyal_input


# 2. Pengujian Homogenitas dengan 3 Nilai Alpha yang Berbeda
alphas = [0.5, 2.0, -1.0]

# Kita siapkan kanvas gambar dengan 3 baris (satu baris untuk setiap nilai alpha)
fig, axs = plt.subplots(3, 1, figsize=(11, 15))

for idx, alpha in enumerate(alphas):
    # Sisi Kiri Rumus: T(alpha * x) -> Input dikali alpha dulu, baru dimasukkan ke sistem
    sisi_kiri = sistem_T(alpha * x)

    # Sisi Kanan Rumus: alpha * T(x) -> Input dimasukkan ke sistem dulu, hasilnya baru dikali alpha
    sisi_kanan = alpha * sistem_T(x)

    # 3. Membuat Plot Perbandingan Overlay (Bertumpuk) untuk Pembuktian
    # Kita gunakan gaya garis berbeda dan transparansi agar terlihat jika berimpit
    axs[idx].stem(
        n, sisi_kiri, linefmt="b-", markerfmt="bo", basefmt="C2-", label="T(α·x)"
    )
    axs[idx].plot(
        n,
        sisi_kanan,
        "r--x",
        linewidth=2,
        markersize=8,
        label="α·T(x)",
        alpha=0.8,
    )

    axs[idx].set_title(
        f"Uji Homogenitas untuk $\\alpha = {alpha}$ (Sumbu Tegak = Amplitudo)"
    )
    axs[idx].grid(True)
    axs[idx].legend(loc="upper right")

    # Cek kesamaan secara matematis di program
    apakah_sama = np.allclose(sisi_kiri, sisi_kanan)
    print(f"Hasil Uji alpha = {alpha:4} -> Apakah T(α·x) == α·T(x)? {apakah_sama}")

plt.tight_layout()
plt.show()
```

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np

# 1. Menyiapkan kembali sinyal x1 (sinus) dan x2 (unit step) dari memori
n = np.arange(0, 31)
x1 = np.sin(0.5 * np.pi * n)
x2 = np.where(n >= 5, 1, 0)


# 2. Mendefinisikan Fungsi Sistem T(x) = 2 * x
def sistem_T(sinyal_input):
    return 2 * sinyal_input


# 3. Pengujian Sisi Kiri: T(x1 + x2)
# Menjumlahkan kedua sinyal dulu, lalu dimasukkan ke sistem
sinyal_jumlah_input = x1 + x2
sisi_kiri = sistem_T(sinyal_jumlah_input)

# 4. Pengujian Sisi Kanan: T(x1) + T(x2)
# Memasukkan masing-masing sinyal ke sistem, lalu hasilnya dijumlahkan
sisi_kanan = sistem_T(x1) + sistem_T(x2)

# 5. Menghitung Selisih (Error) antara Sisi Kiri dan Sisi Kanan
selisih = sisi_kiri - sisi_kanan

# 6. MENAMPILKAN VISUALISASI OUTPUT (5 Plot Bertingkat Sesuai Aturan Tugas)
plt.figure(figsize=(11, 16))

# Plot 1: Input x1[n]
plt.subplot(5, 1, 1)
plt.stem(n, x1, linefmt="b-", markerfmt="bo", basefmt="C2-")
plt.title("1. Sinyal Input $x_1[n]$ (Sinus)")
plt.ylabel("Amplitudo")
plt.grid(True)

# Plot 2: Input x2[n]
plt.subplot(5, 1, 2)
plt.stem(n, x2, linefmt="r-", markerfmt="ro", basefmt="C2-")
plt.title("2. Sinyal Input $x_2[n]$ (Unit Step)")
plt.ylabel("Amplitudo")
plt.grid(True)

# Plot 3: Hasil T(x1 + x2)
plt.subplot(5, 1, 3)
plt.stem(n, sisi_kiri, linefmt="g-", markerfmt="go", basefmt="C2-")
plt.title("3. Hasil Sisi Kiri: $T(x_1[n] + x_2[n])$")
plt.ylabel("Amplitudo")
plt.grid(True)

# Plot 4: Hasil T(x1) + T(x2)
plt.subplot(5, 1, 4)
plt.stem(n, sisi_kanan, linefmt="m-", markerfmt="mo", basefmt="C2-")
plt.title("4. Hasil Sisi Kanan: $T(x_1[n]) + T(x_2[n])$")
plt.ylabel("Amplitudo")
plt.grid(True)

# Plot 5: Selisih antara Kedua Hasil
plt.subplot(5, 1, 5)
plt.stem(n, selisih, linefmt="k-", markerfmt="kx", basefmt="C2-")
plt.title("5. Grafik Selisih (Error) Antara Sisi Kiri dan Sisi Kanan")
plt.xlabel("Sampel (n)")
plt.ylabel("Amplitudo")
plt.ylim([-1, 1])  # Skala diperkecil untuk mempertegas nilai 0
plt.grid(True)

plt.tight_layout()
plt.show()

# Cek numerik nilai selisih maksimum
print("=" * 50)
print(f"Nilai selisih maksimum antar sampel: {np.max(np.abs(selisih))}")
print("=" * 50)
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Menyiapkan input x, x1, dan x2
n = np.arange(0, 31)
x = np.sin(0.5 * np.pi * n)
x1 = np.sin(0.5 * np.pi * n)
x2 = np.where(n >= 5, 1, 0)
alpha = 2.0


# 1. Definisi Sistem 1 (Linier) dan Sistem 2 (Non-Linier)
def T1(sinyal):
    return 2 * sinyal


def T2(sinyal):
    return sinyal**2


# ==========================================
# PROSES PENGUJIAN SISTEM
# ==========================================

# A. Uji Homogenitas
homo_T1_kiri = T1(alpha * x)
homo_T1_kanan = alpha * T1(x)
uji_homo_T1 = np.allclose(homo_T1_kiri, homo_T1_kanan)

homo_T2_kiri = T2(alpha * x)
homo_T2_kanan = alpha * T2(x)
uji_homo_T2 = np.allclose(homo_T2_kiri, homo_T2_kanan)

# B. Uji Additivitas
add_T1_kiri = T1(x1 + x2)
add_T1_kanan = T1(x1) + T1(x2)
uji_add_T1 = np.allclose(add_T1_kiri, add_T1_kanan)

add_T2_kiri = T2(x1 + x2)
add_T2_kanan = T2(x1) + T2(x2)
uji_add_T2 = np.allclose(add_T2_kiri, add_T2_kanan)

# ==========================================
# PRINT OUT HASIL EVALUASI (Output Wajib 1 & 2)
# ==========================================
print("=" * 60)
print("HASIL EVALUASI SISTEM DIGITAL")
print("=" * 60)
print(f"SISTEM 1: T1(x) = 2x")
print(f"  - Memenuhi Homogenitas? {uji_homo_T1}")
print(f"  - Memenuhi Additivitas? {uji_add_T1}")
print(f"  - KESIMPULAN          : SISTEM LINIER")
print("-" * 60)
print(f"SISTEM 2: T2(x) = x^2")
print(f"  - Memenuhi Homogenitas? {uji_homo_T2}")
print(f"  - Memenuhi Additivitas? {uji_add_T2}")
print(f"  - KESIMPULAN          : SISTEM NON-LINIER")
print("=" * 60)

# ==========================================
# VISUALISASI PERBANDINGAN OUTPUT NON-LINIER (T2)
# ==========================================
# Kita plot sistem T2 untuk menunjukkan kegagalannya di grafik laporan
plt.figure(figsize=(12, 8))

plt.subplot(2, 1, 1)
plt.plot(n, homo_T2_kiri, "b-o", label="T2(α·x)")
plt.plot(n, homo_T2_kanan, "r--x", label="α·T2(x)")
plt.title("Uji Homogenitas pada Sistem Non-Linier $T_2(x) = x^2$ (Terlihat grafiknya berbeda)")
plt.grid(True)
plt.legend()

plt.subplot(2, 1, 2)
plt.plot(n, add_T2_kiri, "g-o", label="T2(x1 + x2)")
plt.plot(n, add_T2_kanan, "m--x", label="T2(x1) + T2(x2)")
plt.title("Uji Additivitas pada Sistem Non-Linier $T_2(x) = x^2$ (Terlihat grafiknya berbeda)")
plt.grid(True)
plt.legend()

plt.tight_layout()
plt.show()
```
